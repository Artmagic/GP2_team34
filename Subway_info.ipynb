```python

import requests
import pandas as pd
query = """
[out:json][timeout:90];
area["name"="Москва"]->.searchArea;
(
  node["railway"="station"]["station"="subway"](area.searchArea);
  node["railway"="subway_entrance"](area.searchArea);
);
out body;
"""

url = "https://overpass.kumi.systems/api/interpreter"
print("Запрос Overpass API")
response = requests.get(url, params={"data": query}, timeout=120)
response.raise_for_status()
data = response.json()

stations = []
for el in data["elements"]:
    name = el["tags"].get("name")
    lat = el.get("lat")
    lon = el.get("lon")
    if name and lat and lon:
        stations.append({"name": name, "lat": lat, "lon": lon})

df_metro = pd.DataFrame(stations)
print(f"Станций найдено: {len(df_metro)}")

df_metro.to_csv("moscow_metro.csv", index=False, encoding="utf-8-sig")

```
